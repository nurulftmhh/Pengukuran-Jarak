### Jaccard Similarity

```python
import numpy as np
from scipy.spatial.distance import jaccard
from sklearn.metrics import jaccard_score
```

```python

# Menghitung Jaccard Similarity
jaccard_similarity = []
for i in range(len(trainn)):
    intersection = np.intersect1d(trainn[i], testt)
    union = np.union1d(trainn[i], testt)
    similarity = len(intersection) / len(union)
    jaccard_similarity.append(similarity)

# Menampilkan hasil Jaccard Similarity untuk setiap data train
for i, similarity in enumerate(jaccard_similarity):
    print(f"Jaccard Similarity antara data test dan data train ke-{i+1}: {similarity}")
```

```python
import numpy as np

# Data train
data_train = np.array([
    [3.24, 3.06, 2.79, 2.73, 3.02],
    [2.83, 3.02, 3.02, 3.63, 3.21],
    [2.83, 3.00, 2.54, 3.36, 2.79],
    [3.00, 3.08, 3.02, 3.07, 2.88],
    [3.52, 3.52, 3.31, 3.33, 3.30],
    [2.29, 0.50, 0.16, 0.32, 1.13]
])

# Data test
data_test = np.array([3.1, 3.06, 3.0, 3.23, 2.79])

# Calculate the covariance matrix of the training data
cov_matrix = np.cov(data_train.T)

# Calculate the inverse of the covariance matrix
inv_cov_matrix = np.linalg.inv(cov_matrix)

# Calculate Mahalanobis Distance manually
distances = []
for i in range(len(data_train)):
    diff = data_train[i] - data_test
    dist = np.sqrt(np.dot(np.dot(diff, inv_cov_matrix), diff))
    distances.append(dist)

# Print the Mahalanobis Distances
print("Mahalanobis Distances between data_train and data_test:")
for i, dist in enumerate(distances):
    print(f"Data train ke-{i+1}: {dist}")
```

### Chebyshev distance

```python
from sklearn.metrics import pairwise_distances
# Menghitung jarak Chebyshev antara data train dan data test
distances = pairwise_distances(trainn, testt, metric='chebyshev')

# Menampilkan hasil
print("Jarak Chebyshev antara data train dan data test:")
for i, distance in enumerate(distances):
    print(f"Data train ke-{i+1}: {distance[0]}")
```

### Minkowski distance

```python
# Menghitung jarak Minkowski dengan p=3 (Contoh)
distances = pairwise_distances(trainn, testt, metric='minkowski', p=3)

# Menampilkan hasil
print("Jarak Minkowski antara data train dan data test:")
for i, distance in enumerate(distances):
    print(f"Data train ke-{i+1}: {distance[0]}")
```
